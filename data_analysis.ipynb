```python
import shutil, os
def combine_dataset(dataset_path):

  for folder in os.listdir(dataset_path):
    path = os.path.join(dataset_path, folder)
    for class_fd in os.listdir(path):
      try:
        os.mkdir(os.path.join("/content/dataset",class_fd))
      except Exception as e :
        print(e)



  for folder in os.listdir(dataset_path):
    path = os.path.join(dataset_path, folder)
    for class_fd in os.listdir(path):
      class_path = os.path.join(path,class_fd)
      for img in os.listdir(class_path):
        # print(img)
        shutil.move(os.path.join(class_path,img),os.path.join("/content/dataset",class_fd,img))

combine_dataset("/content/Retinal_Disease_Classification")
```

```python
import os
import matplotlib.pyplot as plt

def get_class_distribution(dataset_path):
    class_distribution = {}
    for class_folder in os.listdir(dataset_path):
        class_path = os.path.join(dataset_path, class_folder)
        if os.path.isdir(class_path):
            class_count = len(os.listdir(class_path))
            class_distribution[class_folder] = class_count
    return class_distribution

def plot_class_distribution(class_distribution):
    classes = list(class_distribution.keys())
    counts = list(class_distribution.values())

    plt.figure(figsize=(10, 5))
    plt.bar(classes, counts, color='blue')
    plt.xlabel('Classes')
    plt.ylabel('Number of Images')
    plt.title('Class Distribution of Image Dataset: ')
    plt.xticks(rotation=45, ha='right')
    plt.show()


def sort_class_distribution(class_distribution):
    sorted_distribution = dict(sorted(class_distribution.items(), key=lambda item: item[1], reverse=True))
    print(sorted_distribution)
    return sorted_distribution

# Replace 'path_to_your_dataset' with the actual path to your image dataset

# for folder in os.listdir('/content/dataset/Retinal_Disease_Classification'):
dataset_path = os.path.join('/content/dataset')
class_distribution = get_class_distribution(dataset_path)
plot_class_distribution(sort_class_distribution(class_distribution))

```

```python
# cleaning the dataset (just once)


import shutil

def remove_dataset(dataset_path, value):
    for class_folder in os.listdir(dataset_path):
      class_path = os.path.join(dataset_path,class_folder)
      if os.path.isdir(class_path):
        if(len(os.listdir(class_path))<value):
          # print("removed "+class_path)
          shutil.rmtree(class_path)

remove_dataset("/content/dataset", 200)


dataset_path = os.path.join('/content/dataset')
class_distribution = get_class_distribution(dataset_path)
plot_class_distribution(sort_class_distribution(class_distribution))
```
